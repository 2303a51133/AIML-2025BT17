Implement Basic Data Structures using Numpy,Pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Part 1: NumPy and pandas Operations
print("Part 1: NumPy and pandas Operations")

# 1. Conversion between NumPy Array and pandas DataFrame
array = np.array([1, 3, 4, 5, 6])
print("\nNumPy Array:")
print(array)

df_from_array = pd.DataFrame(array, columns=['Values'])
print("\npandas DataFrame:")
print(df_from_array)

# 2. Arithmetic Operations
array = np.array([1, 2, 3, 4, 5])
print("\nNumPy Array before multiplication:")
print(array)
array *= 2
print("NumPy Array after multiplication:")
print(array)

df = pd.DataFrame({'Values': [1, 2, 3, 4, 5]})
print("\npandas DataFrame before multiplication:")
print(df)
df['Values'] *= 2
print("pandas DataFrame after multiplication:")
print(df)

# 3. Data Types in List and DataFrame
mixed_list = [1, 2.5, 'hello', True, None]
print("\nTypes in mixed list:")
for element in mixed_list:
    print(f"Element: {element}, Type: {type(element)}")

df_mixed = pd.DataFrame(mixed_list, columns=['Element'])
df_mixed['Type'] = df_mixed['Element'].apply(type)
print("\nDataFrame with types:")
print(df_mixed)

# 4. Arithmetic Operations with NumPy and pandas
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([5, 4, 3, 2, 1])

multiplication_result = np.multiply(array1, array2)
division_result = np.divide(array1, array2)
power_result = np.power(array1, array2)

print("\nNumPy Results:")
print("Multiplication:", multiplication_result)
print("Division:", division_result)
print("Power:", power_result)

df1 = pd.DataFrame({'A': array1})
df2 = pd.DataFrame({'A': array2})

df_multiplication = df1 * df2
df_division = df1 / df2
df_power = df1 ** df2

print("\nPandas Results:")
print("Multiplication:\n", df_multiplication)
print("Division:\n", df_division)
print("Power:\n", df_power)

# 5. Addition of Two Arrays and DataFrame Creation
addition_result = np.add(array1, array2)
df_addition = pd.DataFrame({
    'Array1': array1,
    'Array2': array2,
    'Addition Result': addition_result
})

print("\nAddition Result:")
print(f"Array1 + Array2 = {addition_result}")
print("\nDataFrame with Addition Result:")
print(df_addition)

# 6. NumPy Functions and pandas DataFrame
array = np.array([1, 2, 3, 4, 5])
sin_result = np.sin(array)
log_result = np.log(array)
log2_result = np.log2(array)
exp_result = np.exp(array)

print("\nNumPy Functions Results:")
print("sin:", sin_result)
print("log:", log_result)
print("log2:", log2_result)
print("exp:", exp_result)

df_functions = pd.DataFrame({'Values': array})
df_functions['sin'] = np.sin(df_functions['Values'])
df_functions['log'] = np.log(df_functions['Values'])
df_functions['log2'] = np.log2(df_functions['Values'])
df_functions['exp'] = np.exp(df_functions['Values'])

print("\nPandas DataFrame with Functions:")
print(df_functions)

# Part 2: Plotting with Matplotlib

print("\nPart 2: Plotting with Matplotlib")

data = {
    'People': ['Kiran', 'Arun', 'Vijay', 'Varun'],
    'Age': [25, 30, 35, 40],
    'Height': [145, 151, 165, 173],
    'Weight': [45, 55, 65, 75]
}

df_healthcare = pd.DataFrame(data)
print("\nHealthcare Dataset:")
print(df_healthcare)

plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
plt.scatter(df_healthcare['Age'], df_healthcare['Height'], color='blue')
plt.xlabel('Age')
plt.ylabel('Height (cm)')
plt.title('Age vs Height')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.scatter(df_healthcare['Age'], df_healthcare['Weight'], color='green')
plt.xlabel('Age')
plt.ylabel('Weight (kg)')
plt.title('Age vs Weight')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.bar(df_healthcare['People'], df_healthcare['Height'], color='purple')
plt.xlabel('People')
plt.ylabel('Height (cm)')
plt.title('Height of People')

plt.subplot(2, 2, 4)
plt.hist(df_healthcare['Age'], bins=5, color='orange', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')

plt.tight_layout()
plt.show()

# Part 3: File Handling
print("\nPart 3: File Handling")

# Save to files
df_healthcare.to_csv('data.csv', index=False)
df_healthcare.to_csv('data.tsv', sep='\t', index=False)
df_healthcare.to_excel('data.xlsx', index=False)

# Load from files
print("\nData from TSV file:")
tsv_data = pd.read_csv('data.tsv', sep='\t')
print(tsv_data)

print("\nData from CSV file:")
csv_data = pd.read_csv('data.csv')
print(csv_data)

print("\nData from Excel file:")
excel_data = pd.read_excel('data.xlsx')
print(excel_data)

# Load data from URL
url = 'https://people.sc.fsu.edu/~jburkardt/data/csv/hw_200.csv'
url_data = pd.read_csv(url)
print("\nData from URL:")
print(url_data)
```
